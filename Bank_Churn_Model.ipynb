```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Bank = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Bank%20Churn%20Modelling.csv')
```

```python
Bank = Bank.set_index('CustomerId')
```

```python
Bank.info()
```

```python
Bank['Geography'].value_counts()
```

```python
Bank.replace({'Geography':{'France':2,'Spain':0,'Germany':1}},inplace=True)
```

```python
Bank['Gender'].value_counts()
```

```python
Bank.replace({'Gender':{'Female':1,'Male':0}},inplace=True)
```

```python
Bank['Num Of Products'].value_counts()
```

```python
Bank.replace({'NumOfProducts':{1:0,2:1,3:1,4:1}},inplace=True)
```

```python
Bank['Is Active Member'].value_counts()
```
